# Pacotes necessários

Matplotlib para a criação de gráficos, e os módulos matemáticos.

```python
import matplotlib.pyplot as plt
from math import *
from scipy import *
```

# Definição de funções

Funções necessárias para o cálculo das distribuições e do domínio de cada função.

```python
def binomial_distribuition(n1, N, p):
    valores = []
    for i in range(0, n1+1):
        W = (factorial(N)/(factorial(i)*factorial(N - i)))*((p**(i)))*((1-p)**(N-i))
        valores.append(W)
    return valores

def domain(n1):
    lista = []
    for i in range(0, n1+1):
        lista.append(i)
    return lista

def gaussian_distribuition(n1, N, p):
    valores = []
    sigma = N*(1-p)*p
    for i in range(0, n1+1):
        W = (1/(sqrt(2*pi)*sigma))*exp(-((i-(N*p))**2)/(2*(sigma**2)))
        valores.append(W)
    return valores
```

# Gráficos 

## a) Distribuiçao binomial quando N = 50 e N = 500 com p = 0.5

```python
# Valores fixos

n1 = 50
p = 0.5

# Criação das distribuições e domínios

distribuicao1 = binomial_distribuition(50, 50, p)
dominio1 = domain(50)
distribuicao2 = binomial_distribuition(300, 500, p)
dominio2 = domain(300)

#  Criação d

plt.plot(dominio1, distribuicao1, 'ro', label='N = 50',markersize=4)
plt.plot(dominio2, distribuicao2, 'bo', label='N = 500', markersize=4)
plt.legend(loc='upper right')
plt.xlabel('N_1')
plt.ylabel('Distribuições')
plt.xlim(0, 300)
plt.ylim(0, 0.12)
```

# b) Distribuiçao binomial quando p = 0.1 e p = 0.8 fixando N = 500

```python
n1 = 500
N = 500

distribuicao3 = binomial_distribuition(n1, N, 0.1)
distribuicao4 = binomial_distribuition(n1, N, 0.8)
dominio = domain(n1)

plt.plot(dominio, distribuicao3, 'ro', label = 'p = 0.1', markersize=4)
plt.plot(dominio, distribuicao4, 'bo',label = 'p = 0.8',markersize=4)
plt.legend(loc='upper center')
plt.xlabel('N_1')
plt.ylabel('Distribuições')
plt.xlim(0, 500)
plt.ylim(0, 0.07)
```

## c) Distribuição Gaussiana quando p = 0.1, p = 0.5, p = 0.8 fixando N = 500

```python
n1 = 500
N = 500

distribuicao5 = gaussian_distribuition(n1, N, 0.1)
distribuicao6 = gaussian_distribuition(n1, N, 0.5)
distribuicao7 = gaussian_distribuition(n1, N, 0.8)
dominio = domain(n1)

plt.plot(dominio, distribuicao5, 'ro', label='p = 0.1', markersize=1)
plt.plot(dominio, distribuicao6, 'bo', label='p = 0.5', markersize=1)
plt.plot(dominio, distribuicao7,'go', label ='p = 0.8', markersize=1)

plt.xlabel('N_1')
plt.ylabel('Distribuições')
plt.legend(loc="upper right")
plt.xlim(0, 500)
plt.ylim(0, 0.01)
```
