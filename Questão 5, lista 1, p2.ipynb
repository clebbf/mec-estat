# Pacotes Necessários

```python
import matplotlib.pyplot as plt
from math import *
from scipy import *
```

# Definição das funções (e constante de Boltzmann $\cdot 10^{23}$)

```python
K = 1.380649

def energia(T, E):
    valores = []
    for i in range(1, T+2):
        U = E/(exp(E/(K*i))+1)
        valores.append(U)
    return valores

def calor(T, E):
    valores = []
    for i in range(1, T+2):
        c = ((E**2)/(K*(i**2)))*((exp(E/(K*i)))/((exp(E/(K*i)) +1)**2))
        valores.append(c)
    return valores

def domain(n1):
    valores = []
    for i in range(0, n1+1):
        valores.append(i)
    return valores

def dotted(E):
    valores = []
    for i in range(1, 1002):
        U = E/2
        valores.append(U)
    return valores
```

# Gráficos

## a)  

```python
# Criação das distribuições e domínios

U1 = energia(1000, 20)
U2 = energia(1000, 12)
U3 = energia(1000, 16)
dot1 = dotted(20)
dot2 = dotted(12)
dot3 = dotted(16)
dominio1 = domain(1000)

# Criação do gráfico

plt.plot(dominio1, U1, 'ro', label='U(T), ε = 20',markersize=1)
plt.plot(dominio1, U2, 'bo', label='U(T), ε = 12',markersize=1)
plt.plot(dominio1, U3, 'go', label='U(T), ε = 16',markersize=1)
plt.plot(dominio1, dot1, linestyle='dashed', markersize=0.5, color='black')
plt.plot(dominio1, dot2, linestyle='dashed', markersize=0.5, color='black')
plt.plot(dominio1, dot3, linestyle='dashed',markersize=0.5, color='black')


plt.legend(loc='lower right')
plt.xlabel('T')
plt.ylabel('U')
plt.xlim(0, 1000)
plt.ylim(0, 12)
```

## b)

```python
# Criação das distribuições e domínios

calor1 = calor(1000, 20)
calor2 = calor(1000, 12)
calor3 = calor(1000, 16)
dominio1 = domain(1000)

# Criação do gráfico

plt.plot(dominio1, calor1, 'r-', label='c(T), ε = 20',markersize=1)
plt.plot(dominio1, calor2, 'b-', label='c(T), ε = 12',markersize=1)
plt.plot(dominio1, calor3, 'g-', label='c(T), ε = 16',markersize=1)

plt.legend(loc='upper right')
plt.xlabel('T')
plt.ylabel('c')
plt.xlim(0, 50)
plt.ylim(0, 0.6)
```
