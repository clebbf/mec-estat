# Pacotes necessários

```python
import matplotlib.pyplot as plt
from math import *
from scipy import *
```

# Definição de funções úteis

```python
def binomial_distribuition(n1, N, p):
    valores = []
    for i in range(0, n1+1):
        W = (factorial(N)/(factorial(i)*factorial(N - i)))*((p**(i)))*((1-p)**(N-i))
        valores.append(W)
    return valores

def dominio(n1):
    lista = []
    for i in range(0, n1+1):
        lista.append(i)
    return lista

def gaussian_distribuition(n1, N, p):
    valores = []
    for i in range(0, n1+1):
        W = (factorial(N)/(factorial(i)*factorial(N - i)))*((p**(i)))*((1-p)**(N-i))
        valores.append(W)
    return valores
```

# Gráficos 

## a) Distribuiçao binomial quando N = 50 e N = 500 com p = 0.5

```python
n1 = 50
p = 0.5

distribuicao1 = binomial_distribuition(50, 50, p)
dominio1 = dominio(50)
distribuicao2 = binomial_distribuition(300, 500, p)
dominio2 = dominio(300)


plt.xlim(0, 300)
plt.ylim(0, 0.12)
plt.plot(dominio1, distribuicao1, 'ro', dominio2, distribuicao2, 'bo')
```

# b) Distribuiçao binomial quando p = 0.1 e p = 0.8 fixando N = 500

```python
n1 = 500
N = 500

distribuicao3 = binomial_distribuition(n1, N, 0.1)
distribuicao4 = binomial_distribuition(n1, N, 0.8)
dominio = dominio(n1)

plt.plot(dominio, distribuicao3, 'ro', dominio, distribuicao4, 'bo')
plt.xlim(0, 500)
plt.ylim(0, 0.07)
```
